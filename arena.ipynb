```python
import numpy as np
import matplotlib.pyplot as plt                    
```

## Construímos un array de para a variable independente

```python
print (str(np.pi))
a=np.arange(0,2,0.1)
```

```python
def f(x):
    return x**3-3*x*2+x-1
def g(x):
    return np.cos(np.pi*x)
```

```python
_,ax= plt.subplots()
ax.plot(a,f(a),color="brown",label="polinómica")
ax.plot(a,g(a),color="gold",label="tringonométrica")
legend = ax.legend(loc='lower left', shadow=True, fontsize='small')
legend.get_frame().set_facecolor('C0')
legend.get_frame().set_alpha(0.5)
ax.annotate("Valor Inicial", xy=(1, -5), 
            xytext=(1, -2),
            arrowprops=dict(facecolor='green',width=1,headwidth=10),color='green')
ax.set_title("Gráfica de Función Polinómica e Trigonométrica",color='green')
ax.set_xlabel("x",color="blue")
ax.set_ylabel("y",color="blue")
plt.grid()
plt.show()
```
